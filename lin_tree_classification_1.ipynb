```python
def classify(row):
     if row[1]>160:
            if row[3]>3:
                if row[2] <10:
                    return 510
                else: 
                    return 382
            else:
                if row[2]<40:
                    return 335
                else: 
                    return 312
   
     else:
            if row[2]<25:
                if row [3]>2:
                    return 341
                else: 
                    return 258
            else:
                if row[1]>100:
                    return 285
                else: 
                    return 217
            
```

```python
answer=[np.append(row,classify(row)) for row in dfv ]
answer=np.array(answer)
```
